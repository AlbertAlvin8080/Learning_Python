```python
import tensorflow as tf
```

```python
# indices -> contains the positions of each value.
# values -> contains each individual value to be mapped inside the sparse tensor.
# dense_shape -> contains the actual size of the tensor as a whole.
A = tf.sparse.SparseTensor(
    indices=[[1,1], [3,4]], values=[11,56], dense_shape=[5,6]
)
A
```

```python
tf.sparse.to_dense(A)
```
