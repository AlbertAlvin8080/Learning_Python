```python
import tensorflow as tf
```

```python
# to add some previsibility
tf.random.set_seed(10)

t1 = tf.random.uniform((3,4), minval=10, maxval=99, dtype=tf.int32, seed=5)
t2 = tf.random.uniform((5,3), minval=10, maxval=99, dtype=tf.int32, seed=5)

print(t1)
print(t2)
```

```python
# error: incompatible shapes
# tf.math.add(t1, t2)

# broadcasting: expansion of the dimensions of matrices to make them compatible.
tf.math.add(t1, [2])
```

```python
# ELEMENT WISE
tf.math.multiply(t2, [2])
tf.math.pow(t2, [2])
tf.math.sqrt(tf.cast(t2, dtype=tf.float32), [2])
```

```python
temp = tf.random.normal((3,3), mean=0, stddev=0.5, dtype=tf.float32)
print(temp)
print(tf.math.abs(temp))
```
