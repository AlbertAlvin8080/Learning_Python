```python
import tensorflow as tf
```

```python
A = tf.constant([
    "Franz", "Bonaparta", "Emil", "Sebe"
])
B = tf.constant([
    ["Franz", "Bonaparta"],
    ["Emil", "Sebe"],
    ["Helmuth", "Voss"]
])

A, B
```

```python
tf.strings.join(A, separator="+")
```

```python
# This joins values of rows, not columns
tf.strings.join(B, separator=" ")
```

```python
tf.strings.length(A, unit="BYTE")
tf.strings.length(A, unit="UTF8_CHAR")
```

```python
tf.strings.length(B, unit="UTF8_CHAR")
```

```python
tf.strings.lower(B)
```
