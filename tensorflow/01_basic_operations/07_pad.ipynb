```python
import tensorflow as tf
```

```python
A = tf.random.uniform((4,3), minval=-1, maxval=10, dtype=tf.int32)
B = tf.random.uniform((4,3), minval=-1, maxval=10, dtype=tf.int32)

A, B
```

```python
# [1,1] -> 1 padding at top,  1 padding at bottom
# [2,2] -> 2 padding at left, 2 padding at right
pad = [[1,1], [2, 2]]
# tf.pad(A, paddings=pad, mode="CONSTANT", constant_values=0)
tf.pad(A, paddings=pad, mode="CONSTANT", constant_values=-1)
```

```python
tf.pad(A, paddings=pad, mode="REFLECT")
```

```python
tf.pad(A, paddings=pad, mode="SYMMETRIC")
```
