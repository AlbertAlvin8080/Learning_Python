```python
import tensorflow as tf
import numpy as np
```

```python
t = tf.constant([
    [3, 4, 5],
    [-9, 12, 43],
    [0, 30, 2],
])

print(t)
```

```python
t2 = tf.constant([[2, 3, 4], [5, 6, 7], [8, 9, 10]])

mat_mul = tf.matmul(t, t2, transpose_a=False, transpose_b=False, adjoint_a=False, adjoint_b=False, a_is_sparse=False, b_is_sparse=False, output_type=None, grad_a=False, grad_b=False, name=None)

# also matrix multiplication, but with less options
# mat_mul = t@t2

print(mat_mul)
```

```python
# 4x3
t3 = tf.constant([[2, 3, 4], [5, 6, 7], [8, 9, 10], [9, 9, 9]])

# error: t is (3x3) and t3 is (4x3)
# mat_mul = tf.matmul(t, t3, transpose_a=False, transpose_b=False)

# ok after tranposition
mat_mul = tf.matmul(t, t3, transpose_b=True)

print(mat_mul)
```

```python
# uniform distribution: every value in a certain range has the same chance of being picked at random.
tf.random.uniform((5,5), minval=5, maxval=15)
```

```python
# normal/gaussian distribution: values closer to the mean have more chance of being picked.
# Also, the standard deviation widens or shortens the range of values which are possible to be picked.
tf.random.normal((5,5), mean=0, stddev=1)
```
