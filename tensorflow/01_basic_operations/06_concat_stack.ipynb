```python
import tensorflow as tf
```

```python
A = tf.constant([[7, 8, 9], [10, 5, 0]])
B = tf.constant([[3, -1, 0], [4, 5, 6]])

A, B
```

```python
# concat concatenates a tensor along the specified axis.
tf.concat([A, B], axis=0)
```

```python
tf.concat([A, B], axis=1)
```

```python
# stack creates a new dimension in the specified axis.
tf.stack([A, B], axis=0)
```

```python
tf.stack([A, B], axis=1)
```

```python
tf.random.set_seed(10)

C = tf.random.uniform((4, 3), minval=-1, maxval=10, dtype=tf.int32, seed=5)
D = tf.random.uniform((4, 3), minval=-1, maxval=10, dtype=tf.int32, seed=5)

C, D
```

```python
print(tf.concat([C, D], axis=0))
print(tf.concat([C, D], axis=1))
```

```python
tf.stack([C, D], axis=0)
```

```python
tf.stack([C, D], axis=1)
```

```python
#
tf.stack([C, D], axis=2)
```

```python
# same output
print(tf.stack([C, D], axis=0))
# 4,3 and 4,3 -> 1,4,3 and 1,4,3 -> 2,4,3
print(tf.concat([tf.expand_dims(t, axis=0) for t in [C, D]], axis=0))
```
