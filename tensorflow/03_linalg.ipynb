```python
import tensorflow as tf
```

```python
tf.random.set_seed(10)

t1 = tf.random.uniform((3,3), minval=10, maxval=99, dtype=tf.float32, seed=5)
t2 = tf.random.uniform((3,3), minval=10, maxval=99, dtype=tf.float32, seed=5)

t1, t2
```

```python
# cross product
tf.linalg.cross(t1, t2)
```

```python
# puts each row in a diagonal matrix
tf.linalg.diag(t2)
```

```python
tf.linalg.det(t1), tf.linalg.det(t2)
```

```python
tf.linalg.adjoint(t1)
```

```python
# Inverse matrix
tf.linalg.inv(t1)
```

```python
# Singular Value Decomposition
s, u, v = tf.linalg.svd(t1)

s, u, v
```

```python
# different ways of doing the same thing
x1 = tf.linalg.matmul(t1, t2, transpose_a=True, transpose_b=True)
x2 = tf.linalg.matrix_transpose(t1)@tf.linalg.matrix_transpose(t2)

x1, x2, x1 == x2
```
