  ## Data count  

```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
# df
```

```python
df["Type 2"].value_counts(dropna=False)
```

```python
# Relative Frequency
df["Type 1"].value_counts(normalize=True, dropna=True)
```

```python
df["Type 2"].value_counts(ascending=True, dropna=False)
temp = df["Type 2"].value_counts(ascending=True, dropna=False)["Poison"]

# Treating keys as positions is deprecated. Use Series.iloc[-1] instead.
# temp = df["Type 2"].value_counts(ascending=True, dropna=False)[-1]
temp2 = df["Type 2"].value_counts(ascending=True, dropna=False).iloc[-1] # NaN count in this case

print(temp)
print(temp2)
```

```python
size = df.groupby("Type 1").size()
# size = df.groupby("Type 1").size().add(99) # Testing
size
```

```python
df.groupby("Type 1").count()
```

```python
# Count of all columns across all columns

# df.apply(pd.value_counts) # DEPRECATED
# NOTE: you need to cast all integer series to str because 'str < int' and vice versa cannot happen.
# df.apply(lambda sequence: pd.Series(sequence).value_counts().sort_index())
df.apply(lambda series: pd.Series(series).astype(str).value_counts().sort_index())[:20]
```
