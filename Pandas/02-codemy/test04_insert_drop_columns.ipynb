```python
import pandas as pd
import numpy as np

df = pd.read_csv("pokemon_data.csv")
```

```python
# Inserts at the end
df["Dead"] = [np.nan] * len(df)
df
```

```python
data = [False] * len(df)
df.insert(0, column="Alive", value=data, allow_duplicates=True)
# df.drop(labels="Alive", axis=1, inplace=True) # Removes all columns with "Alive" as label.
df
```

```python
# Returns a new DataFrame because of `inplace=False`) without the columns which have the label "Alive".
dftemp = df.drop(labels="Alive", axis=1, inplace=False) 
dftemp = df.drop(columns="Alive", inplace=False) # Same behavior
dftemp
```

```python
data = [True] * len(df)
# df.assign({"BadPoke":data, "Adopted":True}) # No, **kwargs (keywords) are not dictionaries.
df.assign(BadPoke=data, Adopted=True)[:5] # Returns a copy DF containing the new column.
```
