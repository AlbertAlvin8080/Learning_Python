```python
import pandas as pd

"""
The final objective is to only disnplay customers which should 
be contacted (and format the attributes when possible).
"""

df = pd.read_excel("Customer Call List.xlsx")
# NOTE: Do not drop/replace NaN/Na values this early.
# df.fillna("", inplace=True)
df
```

```python
# TEST

# If there's any NaN/Na, the lambda passed to apply() will not work.
df_copy = df.fillna("", inplace=False) 

# df_copy["Last_Name"].str.find("/")
# df_copy["Last_Name"].str.findall(pat="(/|\.\.\.|_)")

# `pat = "..._/"`
# df_copy["Last_Name"].str.findall(pat="(/|\.\.\.|_)").apply(lambda item: print(type(item)))
mask = df_copy["Last_Name"].str.findall(pat="(/|\.\.\.|_)").apply(lambda item: len(item) != 0)
mask
```

```python
# 1
df.drop_duplicates(inplace=True)
df.drop(columns="Not_Useful_Column", inplace=True)
df
```

```python
# 2

df.rename(columns={"Paying Customer":"Paying_Customer"}, inplace=True)

df["Paying_Customer"] = df["Paying_Customer"].str.replace("Yes", repl="Y", regex=False)
df["Paying_Customer"] = df["Paying_Customer"].str.replace("No", repl="N", regex=False)

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", repl="Y", regex=False)
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", repl="N", regex=False)
# mask = df["Do_Not_Contact"] != "N" # Meaning it should be contacted.

# This filters a lot of observations (rows) and prevents unnecessary processing on them.
# NOTE: Maybe you sould postpone this step in order to understand better the other steps.
# for index in df.index:
#     if df.loc[index, "Do_Not_Contact"] != "N":
#         df.drop(index, inplace=True)

df
```

```python
# 3

# df["Last_Name"].str.lstrip("/...")
# df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("/..._")
# df["Last_Name"] = df["Last_Name"].str.strip("[^a-zA-Z0-9]") # Doesn't work.
df["Last_Name"]
```

```python
# 4

# df["Phone_Number"].astype(dtype=str) # Postponed

# NOTE: The str.replace() method returns NaN for entries that are not strings. To avoid 
# this, we need to ensure all entries are treated as strings before applying the replace method.
# df["Phone_Number"].astype(str).str.replace(pat="[^a-zA-Z0-9]", repl="", regex=True)

# This one also removes N/a (but also filters letters).
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace(pat="\D", repl="", regex=True)
df["Phone_Number"]
```

```python
# 5

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: f"{x[0:3]}-{x[3:6]}-{x[6:10]}")

# This lambda prevents unecessary replacement for invalid values 
# and breaking alreay formatted phone numners.
format_phone = lambda x: f"{x[0:3]}-{x[3:6]}-{x[6:10]}" if len(x) == 10 else x 

df["Phone_Number"] = df["Phone_Number"].apply(format_phone)
df["Phone_Number"]
```

```python
# 6

df_address = df["Address"].str.split(pat=",\s*", n=2, regex=True, expand=True)

df[["Street", "State", "Zip_Code"]] = df_address
df.drop(columns="Address", inplace=True)
df
```

```python
# TEST

# NOTE: You could get rid of duplicated phone numbers, but it still makes 
# sense to be able to contact more than one person using the same phone number, right?
df.drop_duplicates(subset="Phone_Number", inplace=False) # Result if applied.
```

```python
# 7

for index in df.index:
    temp = df.loc[index, ["Phone_Number", "Do_Not_Contact"]]
    # Drops empty phone numbers and people who should not be contacted.
    if temp["Phone_Number"] == "" or temp["Do_Not_Contact"] != "N":
        df.drop(index, inplace=True)

df
```
