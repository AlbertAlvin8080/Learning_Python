```python
import pandas as pd

df1 = pd.read_csv("authors_data_1.csv")
df2 = pd.read_csv("authors_data_2.csv")
```

```python
df1.merge(right=df2, how="inner") # Natural join
```

```python
df1.merge(right=df2, how="inner", on=["author_id"])
```

```python
df1.merge(right=df2, how="inner", on=["author_id", "name"]) 
```

```python
df1.merge(df2, how="left")
```

```python
df1.merge(df2, how="right")
```

```python
df1.merge(df2, how="outer") # full outer join: acts as right + left 
```

```python
df1.merge(df2, how="cross") # cartesian product
```

```python
df_copy = df1.copy()
df_temp = df_copy["name"].str.split(pat=" ", expand=True)
del df_copy["name"]
df_copy.insert(loc=1, column="first_name", value=df_temp.iloc[:, 0])
df_copy.insert(loc=2, column="second_name", value=df_temp.iloc[:, 1])
df_copy

# df_temp.iloc[0] # line
# df_temp.iloc[:, 0] # col
```
