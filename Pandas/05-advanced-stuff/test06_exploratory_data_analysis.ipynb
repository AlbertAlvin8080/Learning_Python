```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# This code uses the modulo (%) operator for string formatting, which is a deprecated approach in Python 3.
# format_number = lambda x: "%.2f" % x
format_number = lambda x: f"{x:.2f}"

# Setting decimal places to 2
pd.set_option("display.float_format", format_number) 

df = pd.read_csv("world_population.csv")
df
```

```python
# df.info()
# df.select_dtypes(include="number") # integers and float values
# df.select_dtypes(include="float") # only float values
df.select_dtypes(include="object") # str and others objects
```

```python
# list(df.groupby("Continent"))

plt.ioff()

metrics = [
    "1970 Population",
    "1980 Population",
    "1990 Population",
    "2000 Population",
    "2010 Population",
    "2015 Population",
    "2020 Population",
    "2022 Population",
]

plotable = df.groupby("Continent")[metrics].mean().transpose()
ax = plotable.plot(figsize=(10, 6), ylabel="Population Mean", xlabel="Year")

# Extract years from the metric names and set them as the new tick labels
years = [metric.split()[0] for metric in metrics]
# years = np.array([metric.split()[0] for metric in metrics])
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

plt.show()
```

```python
plt.ioff()

# NOTE: Boxplot is great for finding outliers. Howerver, this case is special 
# due to the nature of the data we're analysing (Population of countries).
# It makes sense for this type of data to have so many outliers.

# df.boxplot(figsize=(20, 8))
ax = df.select_dtypes(include="number").boxplot(figsize=(20, 8))
ax.tick_params(axis='x', labelsize=12, rotation=35)
ax.tick_params(axis='y', labelsize=12)

# plt.rc('xtick', labelsize=8)
# plt.rc('ytick', labelsize=8)

# Disable scientific notation on the y-axis
ax.get_yaxis().get_major_formatter().set_scientific(False)

plt.show()
```
