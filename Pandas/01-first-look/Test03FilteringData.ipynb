## Filtering Data

```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
# df
```

```python
df.loc[df["Type 1"] == "Flame", "Type 1"] = "Fire"
df.loc[df["Type 1"] == "Fire"][:10].sort_values("#", ignore_index=True)

# df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flame"
# df.loc[df["Type 1"] == "Flame"][:10]
```

```python
import re

# df_grass = df.loc[df["Type 1"] == "Grass"][:10]
df_grass = df.loc[df["Type 1"].str.contains("grass", flags=re.IGNORECASE, regex=True)][:10]
# df_grass.reset_index(drop=True, inplace=True) # Comment this to get the original indexes
df_grass
```

```python
df.loc[df["Type 2"] == "Poison"]
df.loc[df["Type 2"] == "Poison"].reset_index()[:5] # Creates a new column containing the old indexes
df.loc[df["Type 2"] == "Poison"].reset_index(drop=True)[:5] # Doesn't append the old index column
```

```python

# df_poison = df.loc[df["Type 2"] == "Poison"]
# na=False -> removes NaN and N/A by treating them as False for the match
df_poison = df.loc[df["Type 2"].str.contains("poison", flags=re.IGNORECASE, na=False)] 
# na=True  -> NaN and N/A values are treated as True for the match
# df_poison = df.loc[df["Type 2"].str.contains("poison", flags=re.IGNORECASE, na=True)] 

# Option 2 for dealing with NaN and N/A
# df["Type 2"] = df["Type 2"].fillna(False)

# Reduces memory consumption (returns nothing, it modifies the original dataframe)
df_poison.reset_index(drop=True, inplace=True) 
df_poison[:5]

# df["Type 2"].str.contains("poison", flags=re.IGNORECASE)
```
