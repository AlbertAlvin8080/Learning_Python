```python
import pandas as pd
import numpy as np

df = pd.read_csv("GasPricesinBrazil_2004-2019_clean.csv", sep=",")
df.head(1)
```

### Number of records for each product in each region. 

```python
# WRONG: resorts to counting other columns instead of counting the PRODUTO column.
# df.groupby(["PRODUTO", "REGIÃO"]).count()
 
df.groupby(["PRODUTO"])["REGIÃO"].value_counts().to_frame()
```

### How did regular gasoline prices vary in 2018?

```python
partial_1 = df.query("PRODUTO == 'GASOLINA COMUM'")
partial_2 = partial_1.query("ANO == 2018") # WARNING: it's an int64 dtype column, not a string.
partial_2["PREÇO MÉDIO REVENDA"].describe().to_frame()
# partial_2["PREÇO MÉDIO REVENDA"].describe().to_frame().loc["std"]
```

### How did regular gasoline and etanol prices vary in São Paulo in 2018?

```python
partial_1 = df.query("PRODUTO in ['GASOLINA COMUM', 'ETANOL HIDRATADO']")
partial_2 = partial_1.query("ESTADO == 'SAO PAULO'")
partial_3 = partial_2.query("ANO == 2018")
partial_3.groupby("PRODUTO")["PREÇO MÉDIO REVENDA"].describe()
```
