```python
import pandas as pd

# You have to know the number of columns to use `usecols=range(1, 21)`.
# There are 20 columns in this case.
df = pd.read_csv("GasPricesinBrazil_2004-2019.csv", sep=";", usecols=range(1, 21))
df
```

```python
df.query("REGIÃO == 'SUDESTE'")           # Works fine
df.query("PRODUTO == 'ETANOL HIDRATADO'") # Works fine
df.query("UNIDADE DE MEDIDA == 'R$/m3'")  # ERROR: invalid white spaces. Use other filter method.
```

```python
# This approach is simpler, but less performant, because the WHOLE DataFrame will be parsed twice.
filter1 = (df["REGIÃO"] == "SUDESTE") & (df["PREÇO MÉDIO REVENDA"] > 2)
df[filter1]
```

```python
# This approach is more complex, but more performant.
filter1 = df["REGIÃO"] == "SUDESTE"
partial1 = df[filter1]
filter2 = partial1["PREÇO MÉDIO REVENDA"] > 2
partial1[filter2]
```
