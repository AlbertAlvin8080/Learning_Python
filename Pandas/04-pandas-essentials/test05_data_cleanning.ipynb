```python
import pandas as pd
import numpy as np

df = pd.read_csv("GasPricesinBrazil_2004-2019.csv", sep=";", usecols=range(1, 21))
# df.to_csv("test.csv")
df.info()
```

```python
# `errors="coerce"` puts a NaN in the cell when the value cannot be converted.
df["DATA INICIAL"] = pd.to_datetime(df["DATA INICIAL"], errors="coerce")
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"], errors="coerce")

df.info()
```

```python
attributes = ["MARGEM MÉDIA REVENDA", "PREÇO MÉDIO DISTRIBUIÇÃO", "DESVIO PADRÃO DISTRIBUIÇÃO", "PREÇO MÍNIMO DISTRIBUIÇÃO", "PREÇO MÁXIMO DISTRIBUIÇÃO", "COEF DE VARIAÇÃO DISTRIBUIÇÃO"]
for attribute in attributes:
    # `errors="coerce"` puts a NaN in the cell when the value cannot be converted.
    df[attribute] = pd.to_numeric(df[attribute], errors="coerce")
    """
    NOTE: These two approches below do not actually convert the type if there is at least
    one value which cannot be converted to float64 ("-" is the value in this case).
    No, they don't accept `errors="coerce"`.
    """
    # df[column] = df[column].astype(pd.Float64Dtype(), errors="ignore")
    # df[column] = df[column].astype(np.float64, errors="ignore")

# As you can see, the number of non-null values for some columns is now 
# less than the total size of the table.
df.info() 
```

```python
# NOTE: needs previous cells to be axecuted first

# mask = df["DESVIO PADRÃO DISTRIBUIÇÃO"].isnull()
mask = df["DESVIO PADRÃO DISTRIBUIÇÃO"].isna()

# df[mask]        # OK
# df.loc[:, mask] # ERROR
df.loc[mask, :]   # OK
df.loc[mask]      # OK
```

```python
nan_indexes = df[mask].index
df_temp = df.ffill(inplace=False) # ffill takes the previous valid observations.
# df_temp.iloc[nan_indexes]
df_temp.iloc[[1902, 1903]] # These observations now have the same values for previous NaN in 1903.
```

```python
nan_indexes = df[mask].index
df.bfill(inplace=False).iloc[nan_indexes] # bfill takes the next valid obervations.
```
