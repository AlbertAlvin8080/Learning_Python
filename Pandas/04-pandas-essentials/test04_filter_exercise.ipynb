```python
import pandas as pd

df = pd.read_csv("GasPricesinBrazil_2004-2019.csv", sep=";", usecols=range(1, 21))
df.head()
```

## Gas stations from São Paulo or Rio de Janeiro with regular gasoline and average resale price > 2

### V1

```python
filter_1 = (df["ESTADO"] == "SAO PAULO") | (df["ESTADO"] == "RIO DE JANEIRO")
filter_2 = (df["PRODUTO"] == "GASOLINA COMUM")
filter_3 = (df["PREÇO MÉDIO REVENDA"] > 2)
df_result = df[filter_1 & filter_2 & filter_3]
df_result.iloc[[0, 12, 1555]]
```

### V2

```python
states = ["SAO PAULO", "RIO DE JANEIRO"]
# 1
filter_1 = df["ESTADO"].isin(states)
partial_1 = df[filter_1]
# 2
filter_2 = partial_1["PRODUTO"] == "GASOLINA COMUM"
partial_2 = partial_1[filter_2]
# 3
filter_3 = partial_2["PREÇO MÉDIO REVENDA"] > 2
result = partial_2[filter_3]
result.iloc[[0, 12, 1555]]
```

### V3 (Only works for column names which don't have special characters)

```python
df_copy = df.copy()
# Replacing whitespace characters in column names
df_copy.columns = df_copy.columns.str.replace(pat=" ", repl="_")

states = ["SAO PAULO", "RIO DE JANEIRO"]
# 1
partial_1 = df_copy.query("ESTADO in @states")
# 2
partial_2 = partial_1.query("PRODUTO == 'GASOLINA COMUM'")
# 3
result = partial_2.query("PREÇO_MÉDIO_REVENDA > 2")
result.iloc[[0, 12, 1555]]
```
