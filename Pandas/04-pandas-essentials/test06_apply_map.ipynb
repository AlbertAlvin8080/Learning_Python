```python
import pandas as pd
import numpy as np

df = pd.read_csv("GasPricesinBrazil_2004-2019_clean.csv", sep=",")
# df.info()
df.head()
```

### apply vs map vs agg
- **apply (Series)**: Invoke function on values of Series.
- **apply (DataFrame)**: Apply a function along an axis of the DataFrame.
- **map (Series)**: Map values of Series according to an input mapping or function.
- **map (DataFrame)**: Apply a function to a Dataframe elementwise.
- **agg**: Aggregate using one or more operations over the specified axis.

```python
df["PREÇO MÉDIO REVENDA"] = df["PREÇO MÉDIO REVENDA"] * 2
df.head()
```

```python
# NOTE: be careful not to mistake apply() with agg().

# Across all items of an attribute (column)
df["PREÇO MÉDIO REVENDA"].apply(lambda x: x + 10).to_frame().head()
df["PREÇO MÉDIO REVENDA"].map(lambda x: x + 10, na_action="ignore").to_frame().head()
```

```python
def multiply_numbers(x):
    if(pd.api.types.is_numeric_dtype(x)):
        return x + 100
    else:
        return x

# Across all items of the DataFrame
df.apply(multiply_numbers).head()
```
