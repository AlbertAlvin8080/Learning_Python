```python
groupby = df.groupby(["REGIÃO", "PRODUTO"])

# groupby.mean() # TypeError: agg function failed [how->mean,dtype->object]
# groupby.agg(["sum","min"]).head(10) # It also sums strings if you don't specify only numeric attributes.

groupby.agg({"PREÇO MÉDIO REVENDA":"mean", "NÚMERO DE POSTOS PESQUISADOS":"sum"}).head(12)
```

```python
def str_size_comparator(x):
    print(type(x)) # <class 'pandas.core.series.Series'>
    return x.str.len()

groupby = df.groupby(["REGIÃO", "PRODUTO"])
partial = groupby.agg({"PREÇO MÉDIO REVENDA":"mean", "NÚMERO DE POSTOS PESQUISADOS":"sum"})

partial.sort_values(by=["REGIÃO", "PRODUTO"], key=lambda x: x.str.len(), ascending=[False, True])
# partial.sort_values(by=["REGIÃO", "PRODUTO"], key=str_size_comparator, ascending=[False, True])
```
