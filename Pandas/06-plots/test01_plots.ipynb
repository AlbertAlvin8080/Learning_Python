```python
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.DataFrame(np.random.randn(50, 4), columns=["Sun", "Mon", "Tue", "Wed"])
df.head()
```

```python
df.plot(kind="line", figsize=(15,15))
```

```python
df.abs().plot(kind="line", subplots=True, grid=True, colormap="magma")
```

```python
# NOTE: This happens when you forget to reset the indexes.
df.abs().sort_values(by=["Sun"])["Sun"].plot(kind="line", subplots=False, grid=True)
```

```python
df.abs().sort_values(by=["Sun"]).reset_index(drop=True)["Sun"].plot(kind="line", subplots=False, grid=True)
```

```python
df.abs().plot(kind="area", stacked=False, colormap="magma")
```

```python
df.abs().plot.area(stacked=True, colormap="magma")
```

```python
df.plot.box(subplots=False)
```

```python
df.plot.box(subplots=True)
```

```python
df.plot.scatter(x="Sun", y="Mon")
```

```python
# s -> Size
# c -> Thrid parameter or simply a Color
# df.plot.scatter(x="Sun", y="Mon", c="Green", s=30, alpha=0.5)
df.plot.scatter(x="Sun", y="Mon", c="Tue", s=30, alpha=0.5, colormap="magma")
```

```python
df.head(20).plot.bar(stacked=False)
```

```python
df.transpose().plot.bar(legend=False)
```

```python
df.head(20).abs().plot.bar(stacked=True, grid=True)
```

```python
df.abs().transpose().plot.bar(legend=False)
```

```python
df.plot(kind="hist", stacked=False, grid=True, colormap="magma")
```

```python
df.plot(kind="hist", stacked=False, grid=True, colormap="magma", bins=40)
```

```python
df.plot(kind="hist", stacked=True)
```

```python
df_temp = pd.DataFrame(np.random.randn(500, 4), columns=["Sun", "Mon", "Tue", "Wed"])
# C : int or str, optional -> The column label or position for the value of (x, y) point.
df_temp.plot.hexbin(x="Sun", y="Mon", C="Tue", gridsize=50)
del df_temp
```

```python
# kernel density estimation (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random variable. This function uses Gaussian kernels and includes automatic bandwidth determination.
df.plot.kde(title="KDE Plot")
```
