```python
import glob
import math
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image_paths = glob.glob(r"..\..\assets\panorama\*")
print(image_paths)

images = []
for image_path in image_paths:
    temp = cv.imread(image_path)[:,:,::-1]
    images.append(temp)

num_img = len(images)
cols = 3
rows = math.ceil(num_img / cols)

plt.figure()
for i in range(0, num_img):
    plt.subplot(rows, cols, i+1)
    plt.imshow(images[i])
    plt.axis("off")
```

```python
stitcher = cv.Stitcher_create()
status, result = stitcher.stitch(images)

# You may get an error status 1 if the images don't overlap enough. Try with different images.
# Successful settings: You used 400x354 images with 200px > of overlapping.
print(status)
if status == 0:
    plt.figure(figsize=[30,10])
    plt.imshow(result)
    plt.axis("off")
```
