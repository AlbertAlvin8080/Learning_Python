```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
path = os.path.join(os.getcwd(), "..", "assets", "cat2.jpg")
img = cv.imread(path)[:,:,::-1] # BGR to RGB
```

```python
# ADJUSTING BRIGHTNESS
# Matrix used to increment or decrement the brightness of an image.
matrix = np.ones(img.shape, dtype=np.uint8) * 50

brighter = cv.add(img, matrix)
darker = cv.subtract(img, matrix)

plt.figure(figsize=[18,5])
plt.subplot(141); plt.imshow(img); plt.title("Original")
plt.subplot(142); plt.imshow(brighter); plt.title("Brighter")
plt.subplot(143); plt.imshow(darker); plt.title("Darker")
```

```python
# ADJUSTING CONTRAST 
matrix_contrast1 = np.ones(img.shape, dtype=np.float64) * 0.5
matrix_contrast2 = np.ones(img.shape, dtype=np.float64) * 1.5

# np.clip is used here to make sure the multiplied values don't exceed 255 and go all the way around back from 0. 
# That would cause some strange behaviors in the image.
low_contrast = np.uint8(
    np.clip(cv.multiply(np.float64(img), matrix_contrast1), 0, 255)
)
high_contrast = np.uint8(
    np.clip(cv.multiply(np.float64(img), matrix_contrast2), 0, 255)
)

plt.figure(figsize=[18, 5])
plt.subplot(141); plt.imshow(img); plt.title("Original")
plt.subplot(142); plt.imshow(low_contrast); plt.title("Low Contrast")
plt.subplot(143); plt.imshow(high_contrast); plt.title("High Contrast")
```
